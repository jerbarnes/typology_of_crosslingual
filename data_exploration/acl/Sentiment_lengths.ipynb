```python
import numpy as np
import pandas as pd

import sys
sys.path.append("../..")
from utils import utils, model_utils
```

```python
def examples_per_maxlength(info, d, included_langs, tokenizer):
    lengths = [64, 128, 256, 512]
    file_path = info["file_path"]
    lang_name = info["lang_name"]
    dataset = info["dataset"]
    
    # Only for train set
    if lang_name in included_langs and dataset == "train":
        df = pd.read_csv(file_path, header=None)
        df.columns = ["sentiment", "review"]
        d[lang_name] = {}
        d[lang_name]["total"] = df.shape[0] # Total examples
        for l in lengths:
            # How many examples below limit
            n = (df["review"].apply(lambda x: len(tokenizer.encode(x))) <= l).sum()
            d[lang_name][l] = n
            
    return d
```

```python
acl_langs = utils.get_langs("acl")
tokenizer = model_utils.get_tokenizer("xlm-roberta", "sentiment")
```

```python
results = utils.run_through_data("../../data/sentiment/", examples_per_maxlength, 
                                 {}, included_langs=acl_langs, tokenizer=tokenizer)
```

```python
df = pd.DataFrame(results).T.rename_axis("language").reset_index()
df = utils.order_table(df, "acl")
```
